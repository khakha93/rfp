```python
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
dir_path = r"C:\kha\project\rfp\data(sample)"
f_names = [os.path.join(dir_path,f) for f in os.listdir(dir_path)]
df_list = [pd.read_csv(f, sep="\t") for f in f_names]
```

```python
x = [f.split('\\')[-1] for f in f_names]
print(*x, sep='\n')
```

```python
print(*df_list[1].columns, sep='\n')
```

```python
df_list[1].head(3)
```

```python
from gpt import num_tokens_from_string
```

```python
sr_smr = df_list[1].scope.apply(lambda x: num_tokens_from_string(str(x)))

min_val = sr_smr.min()
max_val = sr_smr.max()
mean_val = sr_smr.mean()

bins = range(min_val, max_val + 2)  # 마지막 값 포함 위해 +2

plt.figure(figsize=(25, 5))

ax = sr_smr.hist(bins=bins)
ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.xlim(left=0)
plt.show()
```

```python
sr_smr.describe()
```
